```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Student_scores.csv")
print(df.head())
```

```python
df.info()
```

```python
df.isnull().sum()
```

# Drop Unnamed Column

```python
df=df.drop("Unnamed: 0",axis=1)
```

```python
df.head()
```

# Change weekly study hours

```python
df["WklyStudyHours"]=df["WklyStudyHours"].str.replace("5-Oct","5-10")
```

```python
df.head()
```

# Gender Distribution

```python
plt.figure(figsize=(4,6))
ax=sns.countplot(data=df,x="Gender")
ax.bar_label(ax.containers[0])
plt.title("Gender Distribution")
plt.show()
```

```python
#From the above chart we have anaalysed that:
#The numbers of female in the data is more than the numbers of male.
```

```python
gb = df.groupby("ParentEduc").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb)
```

```python
sns.heatmap(gb,annot=True)
plt.title("Parents Education")
plt.show
```

```python
#From the above chart we have cocluded that the education of the parents have a good imapct on scores
```

```python
gb1 = df.groupby("ParentMaritalStatus").agg({"MathScore":'mean',"ReadingScore":'mean',"WritingScore":'mean'})
print(gb1)
```

```python
sns.heatmap(gb1,annot=True)
plt.title("Parent Marital Status")
plt.show
```

```python
#From the above chart we have cocluded that the Marital Status of the parents have negligible imapct on scores
```

# Distribution of Ethnic Group

```python
groupA = df.loc[(df["EthnicGroup"] == "group A")].count()
groupB = df.loc[(df["EthnicGroup"] == "group B")].count()
groupC = df.loc[(df["EthnicGroup"] == "group C")].count()
groupD = df.loc[(df["EthnicGroup"] == "group D")].count()
groupE = df.loc[(df["EthnicGroup"] == "group E")].count()

l=["group A","group B","group C","group D","group E"]
mlist=[groupA["EthnicGroup"],groupB["EthnicGroup"],groupC["EthnicGroup"],groupD["EthnicGroup"],groupE["EthnicGroup"]]
md={"Group A":groupA["EthnicGroup"],"Group B":groupB["EthnicGroup"],"Group C":groupC["EthnicGroup"],"Group D":groupD["EthnicGroup"],"Group E":groupE["EthnicGroup"]}

plt.pie(mlist, labels = l,autopct = "%1.2f%%")
plt.title("Distribution of Ethnic Group")
plt.show()
print(md)
```

```python
ax=sns.countplot(data=df,x="EthnicGroup")
ax.bar_label(ax.containers[0])
plt.title("Ethnic Groups")
plt.show()
```
